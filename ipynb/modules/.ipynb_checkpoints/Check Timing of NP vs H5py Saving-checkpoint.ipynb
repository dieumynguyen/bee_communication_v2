```python
import h5py
import numpy as np
import time
import matplotlib.pyplot as plt
import sys
import os, shutil
```

```python
X1 = np.arange(-3, 3, 0.01)
X2 = np.arange(-3, 3, 0.01)
```

```python
cmap = lambda t : np.random.normal(0, 1, (X1.shape[0], X2.shape[0]))
```

```python
plt.imshow(cmap(5))
```

```python
def timer(func):
    def inner(*args, **kwargs):
        t0 = time.time()
        func(*args, **kwargs)
        tf = time.time() - t0
        return tf
    return inner
```

## Check Saving Environment

```python
@timer
def write_to_np(history):
    np.save("time_test/test.np", history)
```

```python
@timer
def write_to_h5py(history):
    with h5py.File("time_test/test.h5", "w") as outfile:
        outfile.create_dataset("Concentration", data=history)
```

```python
@timer
def write_to_h5py_preopened(preopened_outfile, history):
    preopened_outfile.create_dataset("Concentration", data=history)
```

```python
NUM_EPOCHS = 50
history = []
times = []
if os.path.exists("time_test"):
    shutil.rmtree("time_test")
    os.makedirs("time_test")
preopened_outfile = h5py.File("time_test/test2.h5", "w")
for epoch_i in range(NUM_EPOCHS):
    sys.stdout.write(f"\rEpoch: {epoch_i+1}/{NUM_EPOCHS}")
    sys.stdout.flush()
    history.append(cmap(epoch_i))
    np_time = write_to_np(history)
    h5_time = write_to_h5py(history)
    if epoch_i > 0:
        del preopened_outfile['Concentration']
    h5_2_time = write_to_h5py_preopened(preopened_outfile, history)
    times.append((np_time, h5_time, h5_2_time))
```

```python
with h5py.File("time_test/test2.h5", "r") as infile:
    print(list(infile.keys()))
```

```python
times = np.array(times)
```

```python
plt.plot(times[:,0], label='np')
plt.plot(times[:,1], label='h5')
plt.plot(times[:,2], label='h5 pre-opened')
plt.legend()
plt.xlabel('Timestep, t')
plt.ylabel("Time (seconds)")
```

## Check Saving Bee Info

```python
def generate_bee():
    bee_info = {
        "x": np.random.uniform(-3, 3),
        "y": np.random.uniform(-3, 3),
        "state": np.random.randint(0, 4),
        "gradient_x": np.random.uniform(-1, 1),
        "gradient_y": np.random.uniform(-1, 1),
        "distance_from_queen": np.random.uniform(-3, 3)
    }
    return bee_info
```

```python
import json
@timer
def write_to_json(history):
    # Save json
    with open('time_test/bee_hist.json', 'w') as outfile:
        json.dump(history, outfile)
```

```python
@timer
def write_to_h5py(history):
    with h5py.File("time_test/bee_hist.h5", "w") as outfile:
        for key, val in history.items():
            outfile.create_dataset(key, data=val)
```

```python
class BeeKeeper():
    def __init__(self):
        self.history = {
            "t" : [],
            "bee_i" : [],
            "x": [],
            "y": [],
            "state": [],
            "gradient_x": [],
            "gradient_y": [],
            "distance_from_queen": []
        }
        
    def update(self, bee_i, bee, t):
        for key, val in bee.items():
            self.history[key].append(val)
            
        self.history['bee_i'].append(bee_i)
        self.history['t'].append(t)
```

```python
NUM_EPOCHS = 50
NUM_BEES = 70
bee_history = []
times = []

if os.path.exists("time_test"):
    shutil.rmtree("time_test")
    os.makedirs("time_test")
    
bee_keeper = BeeKeeper()
for epoch_i in range(NUM_EPOCHS):
    sys.stdout.write(f"\rEpoch: {epoch_i+1}/{NUM_EPOCHS}")
    sys.stdout.flush()
    
    bees = {}
    for bee_i in range(NUM_BEES):
        bee = generate_bee()
        bees[f"bee_{bee_i}"] = bee
        bee_keeper.update(bee_i, bee, epoch_i)
        
    bee_history.append(bees)
    
    json_time = write_to_json(bee_history)
    h5_time = write_to_h5py(bee_keeper.history)
    times.append((json_time, h5_time))
```

```python
times = np.array(times)
plt.plot(times[:,0], label='json')
plt.plot(times[:,1], label='h5')
plt.legend()
plt.xlabel('Timestep, t')
plt.ylabel("Time (seconds)")
```
